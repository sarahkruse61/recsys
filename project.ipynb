```python
import pandas as pd
import os
import json
```

```python
# Dit is die iets kleinere dataset, gaan we denk ik niet gebruiken
# https://www.kaggle.com/code/vatsalmavani/music-recommendation-system-using-spotify-dataset/notebook
path = os.path.join('.', 'data', 'data.csv')
print(f"Je zoekt nu hier: {os.path.abspath(path)}")

df_small = pd.read_csv(path, sep=",", on_bad_lines='skip')
df_small.head()
```

```python
# Dit is de spotify 1 million tracks dataset
# https://www.kaggle.com/datasets/amitanshjoshi/spotify-1million-tracks/data
df_pq = pd.read_parquet('./data/spotify_data.parquet')
df_pq.info()
```

```python
# Dit is de een deel van de dataset van spotify van 1000 random playlists
# https://www.aicrowd.com/challenges/spotify-million-playlist-dataset-challenge/dataset_files
with open('data/mpd.slice.0-999.json') as f:
    data = json.load(f)

df_spot = pd.DataFrame(data['playlists'])
df_spot.head()
```

```python
# Dit is het tweede deel van de dataset van spotify van 1000 random playlists
# https://www.aicrowd.com/challenges/spotify-million-playlist-dataset-challenge/dataset_files
with open('data/mpd.slice.1000-1999.json') as f:
    data = json.load(f)

df_spot2 = pd.DataFrame(data['playlists'])
df_spot2.head()
```

```python
# Hier wordt er per artiest geteld hoe vaak hun naam voorkomt in de kolom tracks.
# Hierdoor hou je bij hoe vaak een bepaalde
# artiest voorkomt in 1000 playlists.
df_tracks1 = df_spot.explode('tracks')
df_tracks1['artist_name'] = df_tracks1['tracks'].apply(lambda x: x['artist_name'])

artist_counts1 = df_tracks1['artist_name'].value_counts()
artist_counts1.head()
```

```python
# Hier wordt er per artiest geteld hoe vaak hun naam voorkomt in de kolom tracks.
# Hierdoor hou je bij hoe vaak een bepaalde
# artiest voorkomt in 1000 playlists. Maar dan hier bij het tweede deel.
df_tracks2 = df_spot2.explode('tracks')
df_tracks2['artist_name'] = df_tracks2['tracks'].apply(lambda x: x['artist_name'])

artist_counts2 = df_tracks2['artist_name'].value_counts()
artist_counts2.head()
```

```python
# Hier worden de twee dataframes van de 1000 playlists gecombineerd
# om totaal te berekenen hoe vaak een artiest voorkomt in 2000
# playlists.
combined = pd.concat([df_spot, df_spot2], ignore_index=True)
combined.head()
```

```python
# Dit zijn alle counts uit 2 files bij elkaar opgeteld. Dus hoe vaak een bepaalde artiest in 2000 playlists voorkomt.
df_tracks = combined.explode('tracks')
df_tracks['artist_name'] = df_tracks['tracks'].apply(lambda x: x['artist_name'])

artist_counts = df_tracks['artist_name'].value_counts()
playlist_counts = artist_counts.reset_index()
playlist_counts.columns = ["artist_name", "playlist_count"]
playlist_counts.head()
```

```python
# Dit is de montly listeners dataset
# https://www.kaggle.com/datasets/meeratif/spotify-top-artists-by-monthly-listeners/data
df_listeners = pd.read_csv('./data/listeners.csv')
df_listeners = df_listeners.rename(columns={"Artist": "artist_name","Listeners": "monthly_listeners"})
df_listeners = df_listeners[["artist_name", "monthly_listeners"]]
#df_listeners[df_listeners["artist_name"] == "Jason Mraz"]
df_listeners.info()
```

```python
# Dit is data die ik geexctract heb  van https://kworb.net/spotify/listeners.html
# Deze heb ik in een csv bestand gezet en hier ingeladen.
df_kwob = pd.read_csv('./data/kwob_data.csv')
df_kwob = df_kwob.rename(columns={"Artist": "artist_name","Listeners": "monthly_listeners"})
df_kwob.info()
```

```python
# Hier merge ik de datasets zodat we één dataframe hebben met alle nodige kolommen
df_merged = df_pq.merge(playlist_counts, on="artist_name", how="left")
df_merged = df_merged.merge(df_kwob, on="artist_name", how="left")

df_artists = df_merged.groupby("artist_name")[["playlist_count","monthly_listeners"]].first().reset_index()
df_artists = df_artists.dropna(subset=["playlist_count", "monthly_listeners"])

df_artists.info()
```
