```python
import pandas as pd
import time

dfs = []

for i in range(1, 11):   # 1 t/m 11 is ruim genoeg voor 24.294 artiesten
    if i == 1:
        url = "https://kworb.net/spotify/listeners.html"
    else:
        url = f"https://kworb.net/spotify/listeners{i}.html"

    try:
        table = pd.read_html(url)[0]
        table["page"] = i
        dfs.append(table)
        print(f"pagina {i} geladen: {table.shape[0]} rijen")
        time.sleep(0.5)
    except Exception as e:
        print(f"pagina {i} mislukt: {e}")
        break

df_all_listeners = pd.concat(dfs, ignore_index=True)
print(df_all_listeners.shape)
```
